```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Generacja danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data)
df = df_raw.copy()  #kopia danych
df
```

```python
df.info()
```

```python
#category lepiej nadaje się do tworzenia wizualizacji niż object
for i in ['size', 'color', 'gender', 'bought']:
    df[i] = df[i].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe().T #statystyki kolumn numerycznych
```

```python
df.describe(include='category').T   #statystyki kolumn kategorycznych
```

#Dane kategoryczne

#LabelEncoder()

```python
#mapowanie targetu
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])  #lub za pomocą jednego polecenia
```

```python
le.classes_ #wyświetla sposób mapowania: no - 0, yes - 1
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])  #cofnięcie mapowania
```

#OneHotEncoder()

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)   #gdy True - macierz rzadka, dla lepszego wglądu
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)   #usunięcie pierwszej kolumny ponieważ 3 nie są potrzebne -> tylko 2
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

#Dane numeryczne

#StandardScaler()

```python
#między pandas as numpy jest różnica w algorytmie std(), mogą występować różnice
#w pandasie estymator std jest nieobciążony a w numpy jest obciążony

print(df['price'])
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")

array = np.array(df['price'])
print(f'Średnia: {array.mean()}')
print(f'Odchylenie standardowe: {array.std():.4f}')
```

```python
#standaryzacja ręczna

def standarize(x):
    return (x - x.mean()) / x.std()

print(standarize(df['price']))
```

```python
from sklearn.preprocessing import scale

scale(df['price'])  #wyniki różnią się bo sklearn sotsuje inny algorytm std, pewnie obciążony
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
df['price'] = scaler.fit_transform(df[['price']])
df['weight'] = scaler.fit_transform(df[['weight']])
df
```
