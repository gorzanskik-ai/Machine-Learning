```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import plotly.express as px

np.random.seed(42)
sns.set(font_scale=1.3)
```

```python
data = make_moons(n_samples=2000, noise=0.25, random_state=42)
X = data[0]
y = data[1]
```

```python
all_data = np.c_[X, y]
df = pd.DataFrame(all_data, columns=['x1', 'x2', 'target'])
df.head()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='RdYlBu', label='train')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='RdYlBu', label='test', marker='x', alpha=0.5)
plt.title('Zbir treningowy i testowy')
plt.legend()
plt.show()
```

Regresja logistyczna

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X, y)
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10, 8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy - dokadno: {classifier.score(X_train, y_train)}')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plot_decision_regions(X_test, y_test, classifier)
plt.title(f'Zbir treningowy - dokadno: {classifier.score(X_test, y_test)}')
plt.show()
```

Drzewo decyzyjne

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth=1)
classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy - dokadno: {classifier.score(X_train, y_train)}')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plot_decision_regions(X_test, y_test, classifier)
plt.title(f'Zbir treningowy - dokadno: {classifier.score(X_test, y_test)}')
plt.show()
```

Przeuczenie 

```python
classifier = DecisionTreeClassifier()   #bez ustalania gbokoci - dopasowanie do granic moliwoci
classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy - dokadno: {classifier.score(X_train, y_train)}')
plt.show()
```
