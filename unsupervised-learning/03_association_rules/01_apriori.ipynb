```python
import pandas as pd
import numpy as np
```

```python
data = {'produkty': ['chleb jajka mleko', 'mleko ser', 'chleb masło ser', 'chleb jajka']}
transactions = pd.DataFrame(data, index=[1, 2, 3, 4])
transactions
```

```python
split = transactions['produkty'].str.split().sum()
uniques = list(set(split))
uniques
```

```python
encoded = pd.DataFrame()
for item in uniques:
    encoded[item] = [1 if item in elements else 0 for elements in data['produkty']]
```

```python
encoded
```

Apriori

```python
from mlxtend.frequent_patterns import apriori, association_rules

supports = apriori(encoded, min_support=0.0000001, use_colnames=True)
supports
```

```python
supports = apriori(encoded, min_support=0.3, use_colnames=True)
supports
```

```python
rules = association_rules(supports, metric='confidence', min_threshold=0.65)
rules = rules.iloc[:, [0, 1, 4, 5, 6]]
rules
```

```python
rules.sort_values(by='lift', ascending=False)
```
